# Training

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CADRE(nn.Module):
    def __init__(self, gene_emb_tensor, num_pathways, emb_dim=200, dropout=0.2):
        super().__init__()
        self.gene_emb = nn.Embedding.from_pretrained(gene_emb_tensor, freeze=True)
        self.pathway_emb = nn.Embedding(num_pathways, emb_dim)
        self.drug_emb = nn.Parameter(torch.randn(num_pathways, emb_dim))

        self.attn_tanh = nn.Linear(emb_dim, emb_dim)
        self.attn_query = nn.Linear(emb_dim, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, gene_indices_batch, pathway_indices):
        B, K = gene_indices_batch.shape  # B=batch size, K=top-k genes
        device = gene_indices_batch.device

        gene_emb = self.gene_emb(gene_indices_batch)              # (B, K, D)
        pathway_emb = self.pathway_emb(pathway_indices).unsqueeze(1)  # (B, 1, D)

        attn_input = torch.tanh(self.attn_tanh(gene_emb + pathway_emb))  # (B, K, D)
        attn_scores = self.attn_query(attn_input).squeeze(-1)            # (B, K)
        attn_weights = F.softmax(attn_scores, dim=-1)                    # (B, K)

        context = torch.sum(attn_weights.unsqueeze(-1) * gene_emb, dim=1)  # (B, D)
        context = self.dropout(context)

        drug_emb = self.drug_emb[pathway_indices]  # (B, D)
        dot = torch.sum(context * drug_emb, dim=1)  # (B,)
        pred = torch.sigmoid(dot)
        return pred
```

```python
from torch.utils.data import Dataset, DataLoader

class GDSCDataset(Dataset):
    def __init__(self, gene_expr_tensor, drug_pathway_tensor, y_tensor):
        self.gene_expr_tensor = gene_expr_tensor
        self.drug_pathway_tensor = drug_pathway_tensor
        self.y_tensor = y_tensor

        num_cells, num_drugs = y_tensor.shape
        self.pairs = [
            (i, j) for i in range(num_cells) for j in range(num_drugs)
            if y_tensor[i, j] != -1
        ]

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        i, j = self.pairs[idx]
        if j >= len(self.drug_pathway_tensor):
            raise IndexError(f"Drug index {j} out of range! Max: {len(self.drug_pathway_tensor) - 1}")
        gene_indices = self.gene_expr_tensor[i]
        pathway_idx = self.drug_pathway_tensor[j]
        label = self.y_tensor[i, j]
        return gene_indices, pathway_idx, label
```

```python
from torch.optim import SGD
from torch.optim.lr_scheduler import OneCycleLR
from sklearn.metrics import roc_auc_score, f1_score

def train(model, train_loader, test_loader, device, epochs=10, max_lr=1e-2):
    model.to(device)
    optimizer = SGD(model.parameters(), lr=max_lr, momentum=0.9)
    scheduler = OneCycleLR(optimizer, max_lr=max_lr, steps_per_epoch=len(train_loader), epochs=epochs)
    criterion = nn.BCELoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        y_true, y_pred = [], []

        for step, (genes, pathways, labels) in enumerate(train_loader):
            genes, pathways, labels = genes.to(device), pathways.to(device), labels.to(device).float()

            pred = model(genes, pathways)
            loss = criterion(pred, labels)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(pred.detach().cpu().numpy())

            if step % 100 == 0:
                auc = roc_auc_score(y_true, y_pred)
                f1 = f1_score(y_true, (torch.tensor(y_pred) > 0.5).int())
                print(f"[Epoch {epoch+1} | Step {step}] Loss: {loss.item():.4f}, AUC: {auc:.4f}, F1: {f1:.4f}")

        print(f"[Epoch {epoch+1} Summary] Avg Loss: {total_loss / len(train_loader):.4f}")
        evaluate(model, test_loader, device)

def evaluate(model, data_loader, device):
    model.eval()
    y_true, y_pred = [], []

    with torch.no_grad():
        for genes, pathways, labels in data_loader:
            genes, pathways = genes.to(device), pathways.to(device)
            pred = model(genes, pathways)
            y_true.extend(labels.numpy())
            y_pred.extend(pred.cpu().numpy())

    auc = roc_auc_score(y_true, y_pred)
    f1 = f1_score(y_true, (torch.tensor(y_pred) > 0.5).int())
    print(f"Test AUC: {auc:.4f}, F1: {f1:.4f}")
```

```python
import torch
import random
from torch.utils.data import DataLoader

# --- Split by cell line index ---
def train_test_split_fixed(y_tensor, gene_expr_tensor, drug_pathway_tensor, test_ratio=0.2, seed=42):
    torch.manual_seed(seed)
    random.seed(seed)

    # --- Split by cell line ---
    num_cells = y_tensor.shape[0]
    num_drugs = y_tensor.shape[1]

    all_cell_idx = list(range(num_cells))
    random.shuffle(all_cell_idx)
    split = int(num_cells * (1 - test_ratio))
    train_cell_idx = all_cell_idx[:split]
    test_cell_idx = all_cell_idx[split:]

    print("y_tensor shape:", y_tensor.shape)
    print("drug_pathway_tensor shape:", drug_pathway_tensor.shape)
    print("First 5 drug IDs in y_tensor:", list(y_tensor.columns)[:5] if hasattr(y_tensor, 'columns') else 'N/A')
    print("First 5 entries in drug_pathway_tensor:", drug_pathway_tensor[:5])

    # --- Mask out drugs with any missing pathways ---
    assert len(drug_pathway_tensor) == num_drugs, "Mismatch between drug response and pathway count"

    # Final aligned outputs
    train_y = y_tensor[train_cell_idx]          # shape [#train_cells, num_drugs]
    test_y = y_tensor[test_cell_idx]            # shape [#test_cells, num_drugs]
    train_genes = [gene_expr_tensor[i] for i in train_cell_idx]
    test_genes = [gene_expr_tensor[i] for i in test_cell_idx]
    train_drugs = drug_pathway_tensor           # shape [num_drugs]
    test_drugs = drug_pathway_tensor

    return train_genes, train_y, train_drugs, test_genes, test_y, test_drugs


def align_y_and_pathways(y_tensor, drug_pathway_tensor):
    # If response matrix has extra drug(s), trim it
    if y_tensor.shape[1] > drug_pathway_tensor.shape[0]:
        print(f"Dropping {y_tensor.shape[1] - drug_pathway_tensor.shape[0]} unmatched drug(s) from y_tensor")
        y_tensor = y_tensor[:, :drug_pathway_tensor.shape[0]]
    elif drug_pathway_tensor.shape[0] > y_tensor.shape[1]:
        print(f"Dropping {drug_pathway_tensor.shape[0] - y_tensor.shape[1]} unmatched pathways")
        drug_pathway_tensor = drug_pathway_tensor[:y_tensor.shape[1]]

    assert y_tensor.shape[1] == drug_pathway_tensor.shape[0], "Still mismatched after fixing!"
    return y_tensor, drug_pathway_tensor


y_tensor_aligned, drug_pathway_tensor_aligned = align_y_and_pathways(y_tensor, drug_pathway_tensor)

# --- Run the split ---
train_gene_expr, train_y_tensor, train_drug_pathways, \
test_gene_expr, test_y_tensor, test_drug_pathways = train_test_split_fixed(
    y_tensor_aligned, gene_expr_tensor, drug_pathway_tensor_aligned, test_ratio=0.2
)

print(train_y_tensor.shape)  # should match [N, D]
print(train_drug_pathways.shape)  # should match [D]

train_dataset = GDSCDataset(train_gene_expr, train_drug_pathways, train_y_tensor)
test_dataset = GDSCDataset(test_gene_expr, test_drug_pathways, test_y_tensor)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128)

# --- Initialize model + optimizer setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = CADRE(
    gene_emb_tensor=gene_emb_tensor,
    num_pathways=len(set(drug_pathway_tensor_aligned.tolist())),
    emb_dim=gene_emb_tensor.shape[1]
)

# --- Train ---
train(model, train_loader, test_loader, device, epochs=10, max_lr=1e-2)
```

# Transformer Encoder

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class CADRETransformer(nn.Module):
    def __init__(self, gene_emb_matrix, num_drugs,
                 emb_dim=200, num_heads=8, num_layers=4,
                 dropout=0.2, pooling='max'):
        super().__init__()

        assert pooling in ['mean', 'max'], "pooling must be 'mean' or 'max'"
        self.pooling = pooling

        # Frozen pretrained gene embeddings
        self.gene_emb = nn.Embedding.from_pretrained(gene_emb_matrix, freeze=True)

        # Transformer encoder setup
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=emb_dim,
            nhead=num_heads,
            dim_feedforward=emb_dim * 4,
            dropout=dropout,
            batch_first=True  # (B, T, D)
        )
        self.encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

        # Learnable drug embeddings
        self.drug_emb = nn.Embedding(num_drugs, emb_dim)

        self.final_dropout = nn.Dropout(dropout)

    def forward(self, gene_indices, drug_ids):
        # gene_indices: [B, 1500] - token indices for top-k expressed genes
        # drug_ids: [B] - integer indices

        gene_embed = self.gene_emb(gene_indices)   # [B, 1500, D]
        encoded = self.encoder(gene_embed)         # [B, 1500, D]

        # Pooling to fixed-size cell line embedding
        if self.pooling == 'mean':
            cell_embed = encoded.mean(dim=1)       # [B, D]
        elif self.pooling == 'max':
            cell_embed = encoded.max(dim=1).values # [B, D]

        cell_embed = self.final_dropout(cell_embed)

        drug_embed = self.drug_emb(drug_ids)       # [B, D]

        # Dot product prediction
        dot = torch.sum(cell_embed * drug_embed, dim=1)  # [B]
        return torch.sigmoid(dot)
```

```python
from torch.utils.data import Dataset

class GDSCDataset(Dataset):
    def __init__(self, gene_expr_tensor, y_tensor):
        self.gene_expr_tensor = gene_expr_tensor
        self.y_tensor = y_tensor

        num_cells, num_drugs = y_tensor.shape
        self.pairs = [
            (i, j) for i in range(num_cells) for j in range(num_drugs)
            if y_tensor[i, j] != -1
        ]

    def __len__(self):
        return len(self.pairs)

    def __getitem__(self, idx):
        i, j = self.pairs[idx]
        gene_indices = self.gene_expr_tensor[i]  # Tensor of shape [1500]
        label = self.y_tensor[i, j]
        return gene_indices, j, label  # j is the drug ID
```

```python
from sklearn.metrics import roc_auc_score, f1_score
from torch.optim import AdamW
from torch.optim.lr_scheduler import OneCycleLR
import torch.nn as nn
import torch

def train(model, train_loader, test_loader, device, epochs=10, max_lr=1e-2):
    model.to(device)
    optimizer = AdamW(model.parameters(), lr=max_lr, weight_decay=1e-2)
    scheduler = OneCycleLR(optimizer, max_lr=max_lr, steps_per_epoch=len(train_loader), epochs=epochs)
    criterion = nn.BCELoss()

    for epoch in range(epochs):
        model.train()
        total_loss = 0
        y_true, y_pred = [], []

        for step, (genes, drug_ids, labels) in enumerate(train_loader):
            genes = genes.to(device)
            drug_ids = drug_ids.to(device)
            labels = labels.float().to(device)

            preds = model(genes, drug_ids)
            loss = criterion(preds, labels)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(preds.detach().cpu().numpy())

            if step % 100 == 0:
                auc = roc_auc_score(y_true, y_pred)
                f1 = f1_score(y_true, (torch.tensor(y_pred) > 0.5).int())
                print(f"[Epoch {epoch+1} | Step {step}] Loss: {loss.item():.4f}, AUC: {auc:.4f}, F1: {f1:.4f}")

        print(f"[Epoch {epoch+1} Summary] Avg Loss: {total_loss / len(train_loader):.4f}")
        evaluate(model, test_loader, device)


def evaluate(model, data_loader, device):
    model.eval()
    y_true, y_pred = [], []

    with torch.no_grad():
        for genes, drug_ids, labels in data_loader:
            genes = genes.to(device)
            drug_ids = drug_ids.to(device)
            preds = model(genes, drug_ids)

            y_true.extend(labels.numpy())
            y_pred.extend(preds.cpu().numpy())

    auc = roc_auc_score(y_true, y_pred)
    f1 = f1_score(y_true, (torch.tensor(y_pred) > 0.5).int())
    print(f"Test AUC: {auc:.4f}, F1: {f1:.4f}")
```
